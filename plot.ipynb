```python
ax = plt.axes (projection="3d")

x_data = np. arange (0, 50, 0.1)
y_data = np.arange (0, 50, 0.1)

X, Y = np.meshgrid (x_data, y_data)
Z = X * Y

ax.plot_surface (X, Y, Z) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate hexagon x and y coordinates
n = 6
r = 1
x = r * np.cos(np.arange(0, 2*np.pi, 2*np.pi/n))
y = r * np.sin(np.arange(0, 2*np.pi, 2*np.pi/n))

# Set the z coordinates to 0
z = np.zeros(n)

# Plot the hexagon
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
